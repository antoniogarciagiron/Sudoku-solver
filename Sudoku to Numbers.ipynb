```python
def sudoku_split_81(figure):     
    h = figure.shape[0] // 9
    w = figure.shape[1] // 9
    subpics = []
    w_i = 1
    w_f = 1
    h_i = 1
    h_f = 1
    for i in range(9):
        h_f += h
        for j in range(9):
            w_f += w
            cropped = figure[h_i:h_f, w_i:w_f]
            #Reduce the frame of each square:
            h_num = cropped.shape[0]
            h_var = (h_num*10) // 100
            w_num = cropped.shape[1]
            w_var = (w_num*10) // 100
            reduced = cropped[h_var:(h-h_var), w_var:(w-w_var)]
            subpics.append(reduced)
            w_i += w
        h_i += h
        w_i = 1
        w_f = 1
    return subpics
```
